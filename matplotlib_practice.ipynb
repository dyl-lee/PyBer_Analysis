```python
%matplotlib inline
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# set x-axis to list of strings for each month
x_axis = ["Jan", "Feb", "Mar","April","May","June","July","Aug","Sept","Oct","Nov","Dec"]

# set y-axis to a list of floats as the total fare in USD accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
```

```python
# recreate plot using object-oriented interface
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
ax = plt.axes()
ax.plot(x_axis, 
        y_axis,
       color="green",
        linewidth=2,
       marker="D")
ax.set_ylabel("Fare ($)")
ax.set_xlabel("Month")
ax.grid()
ax.legend(["Boston"])
ax.set_ylim(0,45)
ax.set_title("PyBer Fare by Month")
plt.savefig("skd.png", format="png")
```

```python
# create plot and add label for legend
plt.plot(
    x_axis, 
    y_axis, 
    marker="*",
    color="blue",
    linewidth=2,
    label="Boston")

# label x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# set y limit between 0 and 45
plt.ylim(0,45)

# create title
plt.title("PyBer Fare by Month")

# Add legend
plt.legend()
```

```python
# create bar graph
plt.bar(x_axis, y_axis,color="green")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer Fare by Month")
plt.legend(["Boston"])
```

```python
# create horizontal bar graph
plt.barh(x_axis, y_axis,color="magenta")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.legend(["Boston"])
plt.gca().invert_yaxis()
```

```python
# create same bar graphs via object oriented interface
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, label='Boston')
```

```python
# create barh via OOI
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, label="Boston",color="cyan")
ax.legend(["Chicago"])
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare by Month")
ax.invert_yaxis()
```

```python
# create scatter plot via plt.plot()
plt.plot(y_axis,x_axis,"o", c="red",label='Chicago')
plt.legend(['Chicago'])
plt.xlabel('Fares ($)')
plt.ylabel('Date')
plt.gca().invert_yaxis()
plt.xlim(0,50)
plt.title('PyBer Fare by Month')
```

```python
# create scatter via plt.scatter()
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis,y_axis,s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s = [i*3 for i in y_axis])
```

```python
# scatter graph using OOI method
fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(y_axis, x_axis, 
           s= [i*5 for i in y_axis], 
           c="skyblue",
           alpha=0.8,
           linewidth=2,
           edgecolors="black")
ax.invert_yaxis()
ax.set_xlim(0,50)
ax.set_xlabel("Fares($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fares by Month")
ax.legend(["Boston"])
plt.savefig("skd2.png", format="png")
```

```python
# pie chart via Matlab 
plt.subplots(figsize = (8,8))
plt.pie(y_axis, labels=x_axis)
```

```python
plt.subplots(figsize = (8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
